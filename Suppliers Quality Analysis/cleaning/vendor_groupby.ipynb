```python
import pandas as pd
import numpy as np
```

```python
# Load your Excel file into a DataFrame
df = pd.read_excel('./Suppliers Quality Analaysis - Cleaned - new.xlsx', sheet_name=None)

defected_items = df['Defected Items']

defected_items
```

```python


# Group the data by Vendor ID
grouped_by_vendor = defected_items.groupby('Vendor ID')

# You can also aggregate the data, for example, summing up defect quantities and downtime minutes
aggregated_data = grouped_by_vendor.agg({
    'Defect Qty': 'sum',
    'Downtime min': 'sum'
}).reset_index()

# Display the aggregated data
aggregated_data
```

```python
# save all dataframes to excel file

with pd.ExcelWriter('Vendor Group By.xlsx') as writer:
    aggregated_data.to_excel(writer, sheet_name='Vendor Group-By', index=False)
```
